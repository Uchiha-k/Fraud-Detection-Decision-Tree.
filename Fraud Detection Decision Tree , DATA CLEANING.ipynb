```python
#Looking for inconsistencies in the data
categorical_cols = df.select_dtypes(include = 'object').columns

for col in categorical_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
    print(f"Unique values: {df[col].nunique()}")
    
```

```python
#handling the ?


df['collision_type'] = df['collision_type'].replace('?', 'Unknown')
df['property_damage'] = df['property_damage'].replace('?', 'Unknown')
df['police_report_available'] = df['police_report_available'].replace('?', 'Unknown')


print(" Question marks handled")
```

```python
#Fix typos on car make/ models

df['auto_make'] = df['auto_make'].replace({
    'Accura' : 'Acura',
    'Suburu' : 'Subaru'
})

df['auto_model'] = df['auto_model'].replace ({
    'Forrestor' : 'Forestor'
})
print(" Car make/model typo fixed")
```

```python
df
```

```python
# Convert Fraud Target to Binary
df['fraud_reported_binary'] = df['fraud_reported'].map({'Y': 1, 'N': 0})
print(f"✅ Fraud conversion: {df['fraud_reported_binary'].value_counts()}")
```

```python
df['fraud_reported']
```

```python
#Days between policy start and incident

df['days_policy_to_incident'] = (df['incident_date'] - df['policy_bind_date']).dt.days
```

```python
df
```

```python
#Policy age at time of claim

df['policy_age_months'] = df['days_policy_to_incident']/30
```

```python
df['total_claim_amount']
```

```python
df['claim_sum'] = df['injury_claim'] + df['property_claim'] + df['vehicle_claim']

df['claim_mismatch'] = df['total_claim_amount'] != df['claim_sum']

print(f"\n  Claim mismatches : {df['claim_mismatch'].sum()}rows")
```

```python
# 6: Check for Impossible Values

print("\n Data Quality Checks:")
print(f"Negative claims: {(df['total_claim_amount'] < 0).sum()}")
print(f"Zero total claims: {(df['total_claim_amount'] == 0).sum()}")
print(f"Claims in future: {(df['incident_date'] > pd.Timestamp.now()).sum()}")
print(f"Negative policy age: {(df['days_policy_to_incident'] < 0).sum()}")
```

```python
#Drop unnecessary columns
df_clean = df.drop(['incident_location'], axis=1)
```

```python
#Fraud verification 

print("\n📋 Final Data Summary:")
print(f"Total rows: {len(df_clean)}")
print(f"Total columns: {len(df_clean.columns)}")
print(f"Missing values: {df_clean.isnull().sum().sum()}")
print(f"Fraud rate: {df_clean['fraud_reported_binary'].mean():.1%}")
```

```python
df_clean.to_csv('insurance_fraud_cleaned.csv', index=False)
print("\n CLEANED DATA SAVED: insurance_fraud_cleaned.csv")
```
