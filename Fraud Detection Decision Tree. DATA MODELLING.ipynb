```python


print("=" * 70)
print(" FRAUD DETECTION FEATURE ANALYSIS REPORT")
print("=" * 70)

# ============================================
# 1. OVERALL FRAUD DISTRIBUTION
# ============================================
print("\n OVERALL FRAUD DISTRIBUTION:")
print(df['fraud_reported'].value_counts())
fraud_rate = df['fraud_reported_binary'].mean()
print(f"\nBase fraud rate: {fraud_rate:.1%}")

# ============================================
# 2. TIME-TO-CLAIM ANALYSIS
# ============================================
print("\n" + "=" * 70)
print(" TIME-TO-CLAIM PATTERNS")
print("=" * 70)

print("\nDays from policy to incident (by fraud status):")
time_analysis = df.groupby('fraud_reported')['days_policy_to_incident'].agg(['count', 'mean', 'median', 'min', 'max'])
print(time_analysis)

quick_fraud = df[df['quick_claim']==1]['fraud_reported_binary'].mean()
slow_fraud = df[df['quick_claim']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate for QUICK claims (<90 days): {quick_fraud:.1%}")
print(f" Fraud rate for SLOW claims (>90 days): {slow_fraud:.1%}")
if slow_fraud > 0:
    print(f" Quick claims are {quick_fraud/slow_fraud:.2f}x more likely to be fraud")

very_quick_fraud = df[df['very_quick_claim']==1]['fraud_reported_binary'].mean()
print(f"\n Fraud rate for VERY QUICK claims (<30 days): {very_quick_fraud:.1%}")

# ============================================
# 3. FINANCIAL PATTERNS
# ============================================
print("\n" + "=" * 70)
print(" FINANCIAL PATTERNS")
print("=" * 70)

print("\nClaim-to-Premium Ratio (by fraud status):")
financial_analysis = df.groupby('fraud_reported')['claim_to_premium_ratio'].agg(['mean', 'median', 'min', 'max'])
print(financial_analysis)

extreme_ratio_fraud = df[df['extreme_claim_ratio']==1]['fraud_reported_binary'].mean()
normal_ratio_fraud = df[df['extreme_claim_ratio']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate for EXTREME ratios (>5x premium): {extreme_ratio_fraud:.1%}")
print(f" Fraud rate for NORMAL ratios: {normal_ratio_fraud:.1%}")
if normal_ratio_fraud > 0:
    print(f" Extreme ratios are {extreme_ratio_fraud/normal_ratio_fraud:.2f}x more likely to be fraud")

high_value_fraud = df[df['high_value_claim']==1]['fraud_reported_binary'].mean()
low_value_fraud = df[df['high_value_claim']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate for HIGH value claims (>$50k): {high_value_fraud:.1%}")
print(f" Fraud rate for NORMAL value claims: {low_value_fraud:.1%}")

# ============================================
# 4. BEHAVIORAL PATTERNS
# ============================================
print("\n" + "=" * 70)
print(" BEHAVIORAL PATTERNS")
print("=" * 70)

# Witnesses
no_witness_fraud = df[df['no_witnesses']==1]['fraud_reported_binary'].mean()
has_witness_fraud = df[df['no_witnesses']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate with NO witnesses: {no_witness_fraud:.1%}")
print(f" Fraud rate WITH witnesses: {has_witness_fraud:.1%}")
if has_witness_fraud > 0:
    print(f" No witnesses = {no_witness_fraud/has_witness_fraud:.2f}x more likely to be fraud")

# Police report
no_police_fraud = df[df['no_police']==1]['fraud_reported_binary'].mean()
has_police_fraud = df[df['no_police']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate with NO police report: {no_police_fraud:.1%}")
print(f" Fraud rate WITH police report: {has_police_fraud:.1%}")

# Authority contact
no_auth_fraud = df[df['no_authority']==1]['fraud_reported_binary'].mean()
has_auth_fraud = df[df['no_authority']==0]['fraud_reported_binary'].mean()
print(f"\n Fraud rate with NO authority contacted: {no_auth_fraud:.1%}")
print(f" Fraud rate WITH authority contacted: {has_auth_fraud:.1%}")

# Multi-vehicle + no witnesses (the suspicious combo)
suspicious_combo_fraud = df[df['multi_vehicle_no_witness']==1]['fraud_reported_binary'].mean()
suspicious_combo_count = df[df['multi_vehicle_no_witness']==1].shape[0]
print(f"\n SUSPICIOUS COMBO (Multi-vehicle + No witnesses):")
print(f"   Cases: {suspicious_combo_count}")
print(f"   Fraud rate: {suspicious_combo_fraud:.1%}")

# ============================================
# 5. RED FLAG SCORE ANALYSIS
# ============================================
print("\n" + "=" * 70)
print(" RED FLAG SCORE DISTRIBUTION")
print("=" * 70)

print("\nFraud rate by red flag count:")
for score in sorted(df['red_flag_score'].unique()):
    count = len(df[df['red_flag_score']==score])
    fraud_rate_score = df[df['red_flag_score']==score]['fraud_reported_binary'].mean()
    fraud_count = df[df['red_flag_score']==score]['fraud_reported_binary'].sum()
    print(f"Red flags = {score}: {fraud_rate_score:>6.1%} fraud rate | {count:>4} total cases | {fraud_count:>3} fraud cases")

# ============================================
# 6. COLLISION TYPE PATTERNS
# ============================================
print("\n" + "=" * 70)
print(" COLLISION TYPE PATTERNS")
print("=" * 70)

collision_fraud = df.groupby('collision_type')['fraud_reported_binary'].agg(['mean', 'count'])
collision_fraud.columns = ['fraud_rate', 'total_cases']
collision_fraud['fraud_rate'] = collision_fraud['fraud_rate'] * 100
print(collision_fraud.sort_values('fraud_rate', ascending=False))

# ============================================
# 7. TOP INSIGHTS SUMMARY
# ============================================
print("\n" + "=" * 70)
print(" KEY INSIGHTS FOR LINKEDIN")
print("=" * 70)

# Calculate lift for top features
insights = []

if slow_fraud > 0:
    quick_lift = quick_fraud / slow_fraud
    insights.append(f"1. Quick claims (<90 days) are {quick_lift:.1f}x more likely to be fraud ({quick_fraud:.1%} vs {slow_fraud:.1%})")

if normal_ratio_fraud > 0:
    ratio_lift = extreme_ratio_fraud / normal_ratio_fraud
    insights.append(f"2. Extreme claim ratios (>5x premium) are {ratio_lift:.1f}x more likely to be fraud ({extreme_ratio_fraud:.1%} vs {normal_ratio_fraud:.1%})")

if has_witness_fraud > 0:
    witness_lift = no_witness_fraud / has_witness_fraud
    insights.append(f"3. Claims with no witnesses are {witness_lift:.1f}x more likely to be fraud ({no_witness_fraud:.1%} vs {has_witness_fraud:.1%})")

# Red flag progression
max_score = df['red_flag_score'].max()
high_score_fraud = df[df['red_flag_score']>=4]['fraud_reported_binary'].mean()
low_score_fraud = df[df['red_flag_score']<=1]['fraud_reported_binary'].mean()
if low_score_fraud > 0:
    insights.append(f"4. Claims with 4+ red flags have {high_score_fraud:.1%} fraud rate vs {low_score_fraud:.1%} for 0-1 flags")

for insight in insights:
    print(f"\n{insight}")

print("\n" + "=" * 70)
print(" ANALYSIS COMPLETE - Ready for modeling!")
print("=" * 70)
```
